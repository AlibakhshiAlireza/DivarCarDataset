```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
```

```python
df = dummies('City',df)
df = dummies('Color',df)
df = dummies('Body_Condition',df)
df = dummies('Engine_Condition',df)
df = dummies('Gearbox',df)
df = dummies('Selling_Method',df)
df = dummies('brandorg',df)
df = dummies('additional info',df)

```
